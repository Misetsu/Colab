# インポート

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
```

# ANDゲート

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], "float32")
X
```

```python
y = np.array([0, 0, 0, 1], "float32")
y
```

```python
plt.figure(figsize=(4, 4))
plt.scatter(X[:,0], X[:, 1], c=y)
plt.show()
```

# モデルの作成

```python
# Frame
model = keras.Sequential()
```

```python
# Add hidden layer
model.add(keras.layers.Dense(1, input_dim=2, activation="sigmoid"))
```

```python
model.summary()
```

# モデルのコンパイル

```python
model.compile(
    # 損失関数(重みとバイアスを更新する時の指標)
    loss="binary_crossentropy",
    # 最適化のアルゴリズム
    optimizer="adam",
    # 評価指標
    metrics=["accuracy"]
)
```

# モデルの学習

```python
history = model.fit(
    X, y,
    epochs=1000,
    verbose=1
)
```

```python
model.evaluate(X, y)
```

# モデルの推論

```python
model.predict(X)
```

# 学習の推移

```python
history.epoch
```

```python
history.history.keys()
```

# 学習結果の可視化

```python
plt.figure(figsize=(8, 2))
plt.plot(history.epoch, history.history["loss"])
plt.title("Loss")
plt.show()
```

```python
plt.figure(figsize=(8, 2))
plt.plot(history.epoch, history.history["accuracy"])
plt.title("Accuracy")
plt.show()
```

```python
x_p = np.linspace(0, 1, 101)
y_p = np.linspace(0, 1, 101)
x_mg, y_mg = np.meshgrid(x_p, y_p)
```

```python
input_array = np.array(list(zip(x_mg.flatten(), y_mg.flatten())))

Z = model.predict(input_array)
Z = Z.reshape(x_mg.shape)

plt.figure(figsize=(4, 4))
plt.contourf(x_mg, y_mg, Z)
plt.scatter(X[:,0], X[:,1], c=y, edgecolors="k")
plt.colorbar()
plt.show()
```

XOR

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], "float32")
X
```

```python
y = np.array([0, 1, 1, 0], "float32")
y
```

```python
plt.figure(figsize=(4, 4))
plt.scatter(X[:,0], X[:, 1], c=y)
plt.show()
```

```python
model1 = keras.Sequential()
```

```python
model1.add((keras.layers.Dense(8, input_dim=2, activation="tanh")))
model1.add((keras.layers.Dense(1, activation="sigmoid")))
model1.summary()
```

```python
model1.compile(
    loss="binary_crossentropy",
    optimizer="adam",
    metrics=["accuracy"]
)
```

```python
history = model1.fit(
    X, y,
    epochs=1000,
    verbose=1
)
```

```python
plt.figure(figsize=(8, 2))
plt.plot(history.epoch, history.history["loss"])
plt.title("Loss")
plt.show()
```

```python
plt.figure(figsize=(8, 2))
plt.plot(history.epoch, history.history["accuracy"])
plt.title("Accuracy")
plt.show()
```

```python
x_p = np.linspace(0, 1, 101)
y_p = np.linspace(0, 1, 101)
x_mg, y_mg = np.meshgrid(x_p, y_p)
```

```python
input_array = np.array(list(zip(x_mg.flatten(), y_mg.flatten())))

Z = model1.predict(input_array)
Z = Z.reshape(x_mg.shape)

plt.figure(figsize=(4, 4))
plt.contourf(x_mg, y_mg, Z)
plt.scatter(X[:,0], X[:,1], c=y, edgecolors="k")
plt.colorbar()
plt.show()
```
